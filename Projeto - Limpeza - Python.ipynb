### <span style='background :PowderBlue'>Carregamento e inspeção inicial</span>

```python
# Importar as bibliotecas Pandas e NumPy.
import pandas as pd
import numpy as np
```

```python
# Carregar o arquivo CSV no DataFrame df.
df = pd.read_csv('audible_uncleaned.csv')

# Inspecionar as primeiras 10 linhas do DataFrame.
df.head(10)
```

O DataFrame possui as seguintes **colunas**:

* `name`: título do audiolivro.
* `author`: autor(es) do audiolivro.
* `narrator`: narrador(es) do audiolivro.
* `time`: duração do audiolivro, em horas e minutos.
* `releasedate`: data de lançamento do audiolivro, no formato dd-mm-aa.
* `language`: idioma do audiolivro.
* `stars`: média das avaliações dadas ao audiolivro, de 0 a 5, seguido do número de avaliações
* `price`: preço em Rupia indonésia.

```python
# Inspecionar a presença de valores nulos, em que colunas ocorrem, e o tipo de dado de cada coluna.
df.info()
```

O DataFrame possui **87489** observações não-nulas por coluna de um total de 87489 observações. À primeira vista, **não há valores nulos** no DataFrame.

Nota-se que os **tipos de dados** de algumas colunas não são os mais adequados. Estes serão corrigidos ao longo do projeto, coluna a coluna. 

`time`, `stars` e `price` deverão ser alterados para um tipo de dado numérico, para facilitar sua manipulação. Será dado preferência ao tipo `float` por ser o tipo mais apropriado para análises de dados exploratórias que possam ser feitas sobre o DataFrame. 

`releasedate` será convertido para `datetime`, no formato dd-mm-aaaa, mais comum no Brasil. As demais colunas (`name`, `author`, `narrator`, `language`), por serem strings, permanecerão como tipo de dado `object`.

```python
# Retornar True para linhas duplicadas.
duplicates = df.duplicated()

# Exibir o número de linhas duplicadas (True), se houver.
print(duplicates.value_counts())
```

**Não há linhas duplicadas** no DataFrame.

<a id="2"></a>
### <span style='background :PowderBlue'>Coluna `name`</span>

Serão removidos possíveis **espaços extras**, caso estes estejam presentes no DataFrame. Por se tratar de um DataFrame longo e com muitas observações únicas, não é factível avaliar coluna a coluna a necessidade de utilizar a função `.strip()`. Portanto, esta será utilizada em todas as colunas de maneira **preventiva**.

```python
# Remover espaços extras.
df['name'] = df['name'].str.strip()
```

Inspecionando o DataFrame, foi possível notar que alguns títulos apresentam espaços antes de dois caracteres especiais: ? e !.

Portanto, serão **removidos os espaços antes de '?' e '!'**.

```python
# Substituir ' ?' por '?', removendo o espaço extra.
df['name'] = df['name'].str.replace(' ?', '?', regex=False) # regex=False não irá considerar ? como uma expressão regular, e sim como caractere.
```

```python
# Substituir ' !' por '!', removendo o espaço extra.
df['name'] = df['name'].str.replace(' !', '!')

# Inspecionar as primeiras 5 linhas do DataFrame até agora.
df.head()
```

Nota-se que nem todos os título estão capitalizados. Porém, **regras de capitalização possuem normas diferentes dependendo do idioma**. Como o DataFrame apresenta diversos idiomas, a capitalização dos títulos será mantida da maneira como está (a qual está correta para o inglês, idioma da maioria dos audiolivros presentes neste DataFrame).

```python
# Inspecionar o número de audiolivros para cada linguagem do DataFrame.
df.language.value_counts()
```

<a id="3"></a>
### <span style='background :PowderBlue'>Coluna `author`</span>

Remover **espaços extras**, se houverem.

```python
# Remover espaços extras.
df['author'] = df['author'].str.strip()
```

**Remover 'Writtenby:'** para que a coluna contenha apenas o nome e sobrenome de cada autor.

```python
# Remover "Writtenby:" substituindo-o por um caractere vazio.
df['author'] = df['author'].str.replace('Writtenby:', '')
```

Adicionar um **espaço entre nome e sobrenome**, os quais não tem espaçamento na coluna.

A linha de código abaixo irá modificar também autores com mais de um sobrenome, com sobrenomes abreviados, e audiolivros que possuem mais de um autor, adicionando espaçamento entre nome e sobrenome para todos.

```python
# Substituir qualquer letra maiúscula por um espaço + a mesma letra utilizando expressões regulares.
df['author'] = df['author'].str.replace('([A-Z]+)', ' \\1', regex=True)
```

Pode-se notar que alguns audiolivros possuem **mais de um autor**.

A coluna `author` será dividida em 4 colunas, cada uma contendo o nome de um autor.

```python
# Separar colunas por autor (author0, author1, author2, e author3) e adicioná-las ao DatFrame.
df = pd.concat([df['name'],
               df['author'].str.split(',', expand = True).add_prefix('author'),
               df.loc[:,['narrator', 'time', 'releasedate', 'language', 'stars', 'price']]], 
               axis = 1)

# Inspecionar as primeiras 5 linhas do DataFrame até agora.
df.head()
```

<a id="4"></a>
###  <span style='background :PowderBlue'>Coluna `narrator`</span>

**Remover espaços extras**, se houverem.

```python
# Remover espaços extras.
df['narrator'] = df['narrator'].str.strip()
```

**Remover 'Narratedby:'** para que a coluna contenha apenas o nome e sobrenome de cada autor.

```python
# Remover "Narratedby::" substituindo-o por um caractere vazio.
df['narrator'] = df['narrator'].str.replace('Narratedby:', '')
```

Assim como na coluna `author`, será adicionado um **espaço entre nome e sobrenome**.

Também serão modificados narradores com mais de um sobrenome, com sobrenomes abreviados, e audiolivros com mais de um narrador, se houverem.

```python
# Substituir qualquer letra maiúscula por um espaço + a mesma letra utilizando expressões regulares.
df['narrator'] = df['narrator'].str.replace('([A-Z]+)', ' \\1', regex=True)
```

Como na coluna `author`, alguns audiolivros também possuem mais de um narrador.

A coluna `narrator` será dividida em 4 colunas, cada uma contendo o nome de um narrador.

```python
# Separar colunas por narrador (narrator0, narrator1, narrator2, e narrator3) e adicioná-las ao DatFrame.
df = pd.concat([df['name'], df['author0'], df['author1'], df['author2'], df['author3'],
               df['narrator'].str.split(',', expand = True).add_prefix('narrator'),
               df.loc[:,['time', 'releasedate', 'language', 'stars', 'price']]], 
               axis = 1)

# Inspecionar as primeiras 5 linhas do DataFrame até agora.
df.head()
```

<a id="5"></a>
###  <span style='background :PowderBlue'>Coluna `time`</span>

**Remover espaços extras**, se houverem.

```python
# Inspecionar as primeiras 5 linhas do DataFrame até agora.
df['time'] = df['time'].str.strip()
```

Para realizar a conversão da coluna `time` para um tipo de dado numérico, o que irá possibilitar a realização de análises numéricas sobre a coluna, iremos transformar os valores, atualmente guardados como 'x hours and x mins', para valores em **minutos**.

Serão extraídos apenas os dígitos correspondentes às horas, e guardados na nova coluna `hours`; o mesmo será feito para os minutos, que serão guardados na nova coluna `minutos`. Então, iremos converter os valores da coluna `hours` para minutos, e somá-los aos valores da coluna `minutos` para obter a duração total de cada audiolivro em minutos. Esta será guardada na coluna original `time`.

```python
# Extrair um ou mais dígitos anteriores à string 'hr' ou 'hrs' e guardá-los na coluna 'hours' utilizando expressões regulares.
# Então, convertê-los para o tipo float para permitir manipulação numérica.
df['hours'] = df['time'].str.extract(r'(\d+) hr?s', expand=False).astype(float)

# Extrair um ou mais dígitos anteriores à string 'min' ou 'mins' e guardá-los na coluna 'hours' utilizando expressões regulares.
# Então, convertê-los para o tipo float para permitir manipulação numérica.
df['minutes'] = df['time'].str.extract(r'(\d+) min?s', expand=False).astype(float)

# Substituir valores nulos por 0 para realizar a conversão para minutos.
df['hours'].replace(np.nan, 0, inplace=True)
df['minutes'].replace(np.nan, 0, inplace=True)

# Substituir os valores na coluna 'time' pelos valores convertidos em minutos.
df['time'] = df['hours'] * 60 + df['minutes']

# Remover as colunas 'hours' e 'minutes', as quais não serão mais utilizadas.
df = df.drop(['hours', 'minutes'], axis=1)

# Inspecionar as primeiras 5 linhas do DataFrame até agora.
df.head()
```

<a id="6"></a>
###  <span style='background :PowderBlue'>Coluna `releasedate`</span>

**Remover espaços extras**, se houverem.

```python
# Remover espaços extras.
df['releasedate'] = df['releasedate'].str.strip()
```

Os valores da coluna, atualmente guardados no formato dd-mm-aa, serão transformados para o **formato dd-mm-aaaa**, mais utilizado no Brasil. A coluna também será convertida para formato de data através da função `pd.to_datetime()`.

```python
# Converter a coluna 'data' para um objetos Pandas datetime.
df['releasedate'] = pd.to_datetime(df['releasedate'], dayfirst=True) # Especificar que o dia virá primeiro.

# Converter as datas para o formato dd-mm-aaaa
df['releasedate'] = df['releasedate'].dt.strftime('%d-%m-%Y')

df.head()
```

<a id="7"></a>
###  <span style='background :PowderBlue'>Coluna `language`</span>

**Remover espaços extras**, se houverem.

```python
# Remover espaços extras.
df['language'] = df['language'].str.strip()
```

A maioria dos idiomas não estão capitalizados, como pode ser visto a seguir:

```python
# Exibir todos os idiomas distintos do DataFrame.
df['language'].unique()
```

Portanto, os idiomas serão **capitalizados** para padronizar a coluna.

```python
# Capitalizar a coluna 'language'.
df['language'] = df['language'].str.title()

# Inspecionar as primeiras 5 linhas do DataFrame até agora.
df.head()
```

<a id="8"></a>
###  <span style='background :PowderBlue'>Coluna `stars`</span>

**Remover espaços extras**, se houverem.

```python
# Remover espaços extras.
df['stars'] = df['stars'].str.strip()
```

Nota-se que a coluna `stars` possui dois tipos de informação: a média das avaliações, e o número de avaliações recebidas pelo audiolivro.

Para permitir a realização de análises sobre os dois tipos de informação, estas serão **separadas em duas colunas**: `stars`, contendo a média das avaliações; e `ratings`, contendo o número de avaliações.

Primeiramente, o número de avaliações será extraído e guardado na nova coluna `ratings`.

```python
# Extrair os dígitos imediatamente anteriores à string 'ratings', e convertê-los para o tipo float.
df['ratings'] = df['stars'].str.extract(r'(\d+) ratings', expand=False).astype(float)

# Substituir valores nulos por 0 para facilitar manipulação numérica.
df['ratings'].replace(np.nan, 0, inplace=True)
```

Então, será guardada apenas a média das avaliações na coluna `stars`.

```python
# Extrair os primeiros dígitos da coluna 'stars' e convertê-los para o tipo float.
df['stars'] = df['stars'].str.extract('(^[0-9].[0-9]*)', expand=False).astype(float)

# Mudar a ordem das colunas para que 'ratings' venha logo após 'stars'.
df = df[['name', 'author0', 'author1', 'author2', 'author3', 'narrator0', 'narrator1', 'narrator2', 'narrator3', 'time', 
         'releasedate', 'language', 'stars', 'ratings', 'price']]

# Inspecionar as primeiras 5 linhas do DataFrame até agora.
df.head()
```

<a id="9"></a>
###  <span style='background :PowderBlue'>Coluna `price`</span>

**Remover espaços extras**, se houverem.

```python
# Remover espaços extras.
df['price'] = df['price'].str.strip()
```

Para possibilitar a realização de análises quantitativas sobre esta coluna, será preciso **convertê-la para um formato numérico**, como `float`.

Primeiramente, serão **removidas as vírgulas** para permitir a manipulação numérica da coluna utilizando Python. Além disso, as linhas que possuem como valor `Free` serão **substituídas por 0**.

```python
# Retirar vírgulas para facilitar manipulação numérica.
df['price'] = df['price'].str.replace(',', '')

# Substituir 'Free' por 0 para permitir manipulação numérica.
df['price'] = df['price'].str.replace('Free', '0')
```

Então, será feita a conversão da coluna para `float`.

```python
# Converter os valores de 'price' para float.
df['price'] = pd.to_numeric(df['price'])

# Inspecionar as primeiras 5 linhas do DataFrame até agora.
df.head()
```
